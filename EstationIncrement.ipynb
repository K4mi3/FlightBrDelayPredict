```python
import pandas as pd
```

```python
df1 = pd.read_csv(r"Input/airports.csv")
df2 = pd.read_excel(r"Input/glossario_de_aerodromo.xls",skiprows=3)
```

```python
df1
```

```python
df_merged = pd.merge(df2, df1, left_on='Sigla OACI', right_on='ident', how='left')
df_merged
```

```python
df_merged.columns
```

```python
import unicodedata

# Função para normalizar os nomes das colunas
def normalize_column_name(col_name):
    # Remove acentos e caracteres especiais
    col_name = unicodedata.normalize('NFKD', col_name).encode('ascii', 'ignore').decode('ascii')
    # Substitui espaços e caracteres especiais por _
    col_name = col_name.replace(" ", "_").replace("-", "_")
    return col_name

# Aplica a normalização a todos os nomes das colunas
df_merged.columns = [normalize_column_name(col) for col in df_merged.columns]

# Exibe os novos nomes das colunas para verificação
print(df_merged.columns)
```

```python
from meteostat import Stations
import math

# Adiciona uma nova coluna ao DataFrame para armazenar os IDs
df_merged['station_id'] = None

for idx, row in df_merged.iterrows():
    if row['latitude_deg'] is None or math.isnan(row['latitude_deg']):
        print(f"Coordenadas inválidas na linha {idx}")
    else:
        # Busca estações próximas
        stations = Stations()
        stations = stations.nearby(lat=row['latitude_deg'], lon=row['longitude_deg'])
        
        if stations._data is not None and not stations._data.empty:
            # Coleta o id da primeira estação
            station_id = stations._data.index[0]
            print(f"ID da estação para a linha {idx}: {station_id}")
            
            # Adiciona o ID ao DataFrame
            df_merged.at[idx, 'station_id'] = station_id
        else:
            print(f"Nenhuma estação encontrada próxima às coordenadas na linha {idx}")
```
