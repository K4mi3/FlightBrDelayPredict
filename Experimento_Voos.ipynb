```python
import os
import pandas as pd
from datetime import datetime
from unidecode import unidecode
from meteostat import Point, Hourly
```

```python
pasta_2023 = 'Input/Voos/2023'
pasta_2024 = 'Input/Voos/2024'

df_voos = []

def ler_arquivos_da_pasta(pasta):
    for arquivo in os.listdir(pasta):
        if arquivo.endswith('.csv'):
            caminho_arquivo = os.path.join(pasta, arquivo)
            df = pd.read_csv(caminho_arquivo, delimiter=";")
            df_voos.append(df)


ler_arquivos_da_pasta(pasta_2023)
ler_arquivos_da_pasta(pasta_2024)
df_final = pd.concat(df_voos, ignore_index=True)
```

```python
df_aeroporto = df_filtrado.rename(columns={'gps_code': 'Sigla ICAO Aeroporto'})
```

```python
df_voo = df_final.merge(
    df_aeroporto,
    left_on='Sigla ICAO Aeroporto Origem',
    right_on='Sigla ICAO Aeroporto',
    suffixes=('', '_origem')
)
df_voo = df_voo.merge(
    df_aeroporto,
    left_on='Sigla ICAO Aeroporto Destino',
    right_on='Sigla ICAO Aeroporto',
    suffixes=('', '_destino')
)

```

```python
df_voo = df_voo.drop(columns=[ 'Sigla ICAO Aeroporto_destino'])
```

```python

df_voo['Partida Real'] = pd.to_datetime(df_voo['Partida Real'], errors='coerce')
df_voo['Chegada Real'] = pd.to_datetime(df_voo['Chegada Real'], errors='coerce')
```

```python
df_voo.columns = df_voo.columns.str.replace(' ', '_')
```

```python
import unicodedata

# Função para normalizar os nomes das colunas
def normalize_column_name(col_name):
    # Remove acentos e caracteres especiais
    col_name = unicodedata.normalize('NFKD', col_name).encode('ascii', 'ignore').decode('ascii')
    # Substitui espaços e caracteres especiais por _
    col_name = col_name.replace(" ", "_").replace("-", "_")
    return col_name

# Aplica a normalização a todos os nomes das colunas
df_voo.columns = [normalize_column_name(col) for col in df_voo.columns]

# Exibe os novos nomes das colunas para verificação
print(df_voo.columns)
```

```python
from meteostat import Point, Hourly

def ObterDadosMeterologicos(latitude,longitude,start,end):
    
    point_st = Point(latitude, longitude)
    TempoDia = Hourly(point_st,start=start,end=end)
    return TempoDia.fetch()
    
```

```python
df_voo = df_voo.dropna(subset=['Partida_Real', 'Chegada_Real'])
```

```python
df_dados = []
for row in df_voo.itertuples(index=True):
    #print(f"longitude={row.longitude_deg}, latitude={row.latitude_deg},start={row.Partida_Real},end={row.Chegada_Real}")
    
    Retorno_dia = ObterDadosMeterologicos(longitude=row.longitude_deg,latitude=row.latitude_deg,start=row.Partida_Real,end=row.Chegada_Real)
    if not Retorno_dia.empty:
        
        df_voo.at[row.Index, 'temp'] = Retorno_dia['temp'].iloc[0] 
        df_voo.at[row.Index, 'dwpt'] = Retorno_dia['dwpt'].iloc[0]
        df_voo.at[row.Index, 'rhum'] = Retorno_dia['rhum'].iloc[0]
        df_voo.at[row.Index, 'prcp'] = Retorno_dia['prcp'].iloc[0]
        df_voo.at[row.Index, 'snow'] = Retorno_dia['snow'].iloc[0]
        df_voo.at[row.Index, 'wdir'] = Retorno_dia['wdir'].iloc[0]
        df_voo.at[row.Index, 'wspd'] = Retorno_dia['wspd'].iloc[0]
        df_voo.at[row.Index, 'wpgt'] = Retorno_dia['wpgt'].iloc[0]
        df_voo.at[row.Index, 'pres'] = Retorno_dia['pres'].iloc[0]
        df_voo.at[row.Index, 'tsun'] = Retorno_dia['tsun'].iloc[0]
        df_voo.at[row.Index, 'coco'] = Retorno_dia['coco'].iloc[0]
      
        
        
        
```
